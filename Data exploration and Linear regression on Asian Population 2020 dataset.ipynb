```python
import pandas as pd

df = pd.read_csv ('AsiaPopulation2020.csv')


#x=df['FertRate']
#y=df['YearlyChange']


# starting from a random pint say (0,0)
t1 = 0
t0 = 0

alp = 0.001  # The learning Rate
itr = 400  # The number of iterations to perform gradient descent

n = float(len(x)) # Number of elements in x or y

# Performing Gradient Descent 
for i in range(itr): 
    y_pred = t1*x + t0  # The origin line function which predicts value of Y
    sigma_for_t1 = (1/n) * sum(x * (y_pred - y))  # sigma of((1/n)(x(h(x(i)-y(i))))) from i=0 to m
    sigma_for_t0 = (1/n) * sum(y_pred - y)  # sigma of((1/n)((h(x(i)-y(i))))) from i=0 to m
    t1 = t1 - alp * sigma_for_t1  # Update m
    t0 = t0 - alp * sigma_for_t0  # Update c
    #t1 and t2 are given at last because the update must be done simultaneously
    
print ('slope:t1 is',t1)
print('intercept t0 is',t0)
```

```python
y_pred=t1*x + t0
print('the predicted y values are:')
print(y_pred)
```
